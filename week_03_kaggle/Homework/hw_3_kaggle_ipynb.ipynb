(Дополнительно) Если вы нашли какие-то ошибки в данных или выбросы, то можете их убрать. Тут можно поэксперементировать с обработкой данных как угодно, но не за баллы.

```python
def remove_outliers(df, columns, method='IQR'):
    df_clean = df.copy()

    if method == 'IQR':
        for col in columns:
            Q1 = df_clean[col].quantile(0.25)
            Q3 = df_clean[col].quantile(0.75)
            IQR = Q3 - Q1
            lower_bound = Q1 - 1.0 * IQR
            upper_bound = Q3 + 1.0 * IQR
            df_clean = df_clean[(df_clean[col] >= lower_bound) & (df_clean[col] <= upper_bound)]

    return df_clean

data_clean = remove_outliers(data, num_cols)

fig, axes = plt.subplots(len(num_cols), 1, figsize=(6, 3 * len(num_cols)))

for i, col in enumerate(num_cols):
    bins = np.histogram_bin_edges(data[col].dropna(), bins=20)

    axes[i].hist(data[col], bins=bins, alpha=0.5, color='red', label='Before (Raw Data)')
    axes[i].hist(data_clean[col], bins=bins, alpha=0.5, color='blue', label='After (Cleaned Data)')

    axes[i].set_title(f'Comparison of {col} (Before vs After)')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Frequency')
    axes[i].legend()

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, OneHotEncoder, LabelEncoder
from sklearn.pipeline import make_pipeline
from sklearn.compose import ColumnTransformer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score


X_train, X_valid, y_train, y_valid = train_test_split(data_clean.drop(target_col, axis=1), data_clean[target_col], test_size=0.2, random_state=42)

preprocessor = ColumnTransformer([
    ('num', StandardScaler(), num_cols),
    ('cat', OneHotEncoder(handle_unknown='ignore'), cat_cols)
])

pipeline = make_pipeline(preprocessor, LogisticRegression(max_iter=1000, solver='liblinear'))

param_grid = {
    'logisticregression__C': [100, 10, 1, 0.1, 0.01, 0.001],
    'logisticregression__penalty': ['l1', 'l2'],
    'logisticregression__solver': ['liblinear', 'saga'],
    'logisticregression__class_weight': [None, 'balanced']
}

grid_search = GridSearchCV(pipeline, param_grid, cv=5, scoring='roc_auc', n_jobs=-1, refit=True)
grid_search.fit(X_train, y_train)

# Best hyperparameters
best_params = grid_search.best_params_
print(f"Best parameters: {best_params}")

# Evaluate on validation set
y_pred_proba = grid_search.best_estimator_.predict_proba(X_valid)[:, 1]
roc_auc = roc_auc_score(y_valid, y_pred_proba)
print(f"ROC-AUC on validation: {roc_auc}")
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import roc_auc_score
from catboost import CatBoostClassifier


X_train, X_valid, y_train, y_valid = train_test_split(data_clean.drop(target_col, axis=1), data_clean[target_col], test_size=0.2, random_state=42)

model = CatBoostClassifier(cat_features=cat_cols, verbose=0, random_state=42)
model.fit(X_train, y_train)

y_pred_proba = model.predict_proba(X_valid)[:, 1]
roc_auc = roc_auc_score(y_valid, y_pred_proba)
print(f"Baseline ROC-AUC: {roc_auc:.4f}")
```

## Запустим перебор параметров обучая модель на GPU

```python
import torch
from catboost import CatBoostClassifier, Pool

task_type = "GPU" if torch.cuda.is_available() else "CPU"
print(f"Using {task_type} for training.")

X_train, X_valid, y_train, y_valid = train_test_split(data_clean.drop(target_col, axis=1), data_clean[target_col], test_size=0.2, random_state=42)

train_pool = Pool(X_train, label=y_train, cat_features=cat_cols)
valid_pool = Pool(X_valid, label=y_valid, cat_features=cat_cols)

model = CatBoostClassifier(
    task_type=task_type,
    cat_features=cat_cols,
    verbose=0,
    random_seed=42
)

param_grid = {
    'n_estimators': [100, 500, 1000],
    'learning_rate': [0.1, 0.05, 0.01],
    'depth': [4, 6, 8],
    'l2_leaf_reg': [1, 3, 5],
    'boosting_type': ['Ordered', 'Plain'],
    'bootstrap_type': ['Bayesian', 'Bernoulli', 'MVS'],
    'random_strength': [1, 5, 10],
    'bagging_temperature': [0, 1, 5]
}

grid_search = GridSearchCV(
    model,
    param_grid,
    cv=5,
    scoring='roc_auc',
    n_jobs=-1
)
grid_search.fit(X_train, y_train)

# Лучшие параметры
best_params = grid_search.best_params_
print(f"Best parameters: {best_params}")

# Оценка лучшей модели на validation
best_model = grid_search.best_estimator_
y_pred_proba = best_model.predict_proba(X_valid)[:, 1]
roc_auc = roc_auc_score(y_valid, y_pred_proba)
print(f"Best ROC-AUC on validation: {roc_auc:.4f}")
```
